```python
def get_ids(csv_name):
    track_ids = []
    chart = pd.read_csv(csv_name)
    for i in range(1,len(chart)):
        track = chart['Unnamed: 4'][i].split("/")[4]
        track_ids.append(track)
    return track_ids
```
