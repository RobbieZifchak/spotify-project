```python
import requests
from bs4 import BeautifulSoup as BS
import re
```

```python
url = 'https://spotifycharts.com/regional/global/daily/2019-12-21'
```

```python
results = requests.get(url)
```

```python
results.status_code
```

```python
soup = BS(results.content, 'html.parser')
```

```python
soup.prettify()
```

```python
dirty_songs = soup.find_all('td', class_ = "chart-table-track")
```

```python
songs = []
for ele in dirty_songs:
    songs.append(ele.get_text(separator = ' '))
```

```python
songs[0].split('\n')
```

```python
mylist = []
for song in songs:
    mylist.append(song.strip().split('\n')) 
```

```python
mylist[0][1]
```

```python
i=0
name=[]
artist=[]
for thing in mylist:
        name.append(mylist[i][0])
        artist.append(mylist[i][1])
        i+=1
```

```python
artist[0]
```

```python
#name of songs
name
```

```python
new_artist = []
for item in artist:
    maybe=item.replace('by','')
    new_artist.append(maybe)
```

```python
#name of artist
new_artist
```

```python
new_new_artist = []
for item in new_artist:
    maybe=item.replace(' %','')
    new_new_artist.append(maybe)
```
